```python
import pandas as pd
```

```python
df = pd.read_csv("data/participationform - Respostas.csv")
df.columns = ["date", "name", "email", "instituicao", "degree", "nan", "palestras"]
df.head()
```

```python
palestras = {
    '3-14':  'Os impactos da Matemática e da Ciência de Dados no Mercado Financeiro',
    '3-1530': 'Vizagrams: um mergulho na teoria de visualização de dados e uma proposta de avanço',
    '3-17':  'Dou-lhe uma, dou-lhe duas... vendido para aquele robô!',
    '4-14':  'Não importa o que você faça, se souber Matemática fará melhor!',
    '4-1530': 'Estamos passando por uma instabilidade em nosso sistema',
    '4-17':  'Princípio da Casa dos Pombos',
    '5-14':  'Teoremas de Existência: da Fórmula de Bhaskara à SVD',
    '5-16':  'Olimpíadas de matemática no Brasil',
}

listas_presenca = {pid: [] for pid in palestras.keys()}
```

Agregar inscrições por palestras

```python
# Percorre cada linha do CSV
c = 0
for _, row in df.iterrows():
    name = str(row["name"]).strip()
    instituicao = str(row["instituicao"]).strip()
    inscricao = str(row["palestras"])
  
    for item in inscricao.split("|"):
        if ":" in item:
            pid, _ = item.split(":")
        else:
            pid = item.strip()

        if pid in palestras:
            listas_presenca[pid].append({
                "name": name,
                "instituicao": instituicao
            })
            c+=1
            
print(c)
```

Gerar CSV

```python
for pid, inscritos in listas_presenca.items():
    df_palestra = pd.DataFrame(inscritos)
    df_palestra.to_csv(f"data/{palestras[pid].replace(' ', '_').replace(":", "")}.csv", index=False, encoding="utf-8-sig")
```

Total de inscritos

```python
c = 0
for pid, inscritos in listas_presenca.items():
    print(f"{palestras[pid]} - {len(inscritos)} inscritos")
    c+=len(inscritos)
print(c)
```
